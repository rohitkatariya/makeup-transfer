```python
import dlib
import cv2
```

```python
import face_recognition
image = face_recognition.load_image_file("./data/input/color_human.jpg")
face_locations = face_recognition.face_locations(image)
```

```python
face_locations
```

```python
face_landmarks_list = face_recognition.face_landmarks(image)
```

```python
face_locations
```

```python
color = (0, 255, 0)
color2 = (0,  0,255)
# Line thickness of 9 px
thickness = 1
```

```python
face_locations[0]
```

```python
image_landmarked = image.copy()
this_face_location = face_locations[0]
cv2.rectangle(image_landmarked , (this_face_location[1],this_face_location[0]), (this_face_location[3],this_face_location[2]), color, thickness)            
for (z,z_landmarks) in face_landmarks_list[0].items():
    print(z)
    prev_landmark = None
    for this_landmark in z_landmarks:
        if prev_landmark is not None:
            cv2.line(image_landmarked, prev_landmark, this_landmark, color, thickness)
            cv2.circle(image_landmarked, this_landmark, 1, color2, thickness)
            
            print(prev_landmark,this_landmark)
        prev_landmark = this_landmark
```
